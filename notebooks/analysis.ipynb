```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

### Champions

```python
with open('./data/champion.json') as f:
  champion_contents = f.read()

champion_json = json.loads(champion_contents)

champions = []
for _, champ_data in champion_json['data'].items():
  champions.append(champ_data)

champion_df = pd.DataFrame(champions)
champion_df

```

#### Limpando os dados

```python
champion_df = champion_df.drop(columns=['blurb', 'title', 'image', 'id'])

champion_df
```

#### From To -> Tags

```python
champion_df['tags'] = champion_df['tags'].apply(lambda x: x[0])
champion_df
```

```python
champion_df.loc[champion_df['tags'] == 'Fighter', 'tags'] = 'JUNGLE'
champion_df.loc[champion_df['tags'] == 'Tank', 'tags'] = 'TOP'
champion_df.loc[champion_df['tags'] == 'Mage', 'tags'] = 'MID'
champion_df.loc[champion_df['tags'] == 'Assassin', 'tags'] = 'BOT'
champion_df.loc[champion_df['tags'] == 'Marksman', 'tags'] = 'BOT'
champion_df.loc[champion_df['tags'] == 'Support', 'tags'] = 'SUP'

champion_df
```

#### Analisando um Champion de exemplo

```python
poppy = champion_df.loc[champion_df['name'] == 'Poppy']
poppy_stats = poppy['stats'].to_dict()
poppy_stats
```
